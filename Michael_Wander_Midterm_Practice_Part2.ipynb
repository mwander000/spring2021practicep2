## Dataset: Dry Bean Dataset from UCI Machine Learning Repository

Multi-class classification problem

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import requests
import zipfile
import io
```

## Preprocessing

```python
zfile = zipfile.ZipFile(io.BytesIO(requests.get('https://archive.ics.uci.edu/ml/machine-learning-databases/00602/DryBeanDataset.zip').content))
worksheet = zfile.read('DryBeanDataset/Dry_Bean_Dataset.xlsx')
```

```python
df = pd.read_excel(worksheet)
df.head()
```

```python
df.info()
```

There is no missing data. There are seven unique classes:

```python
df['Class'].unique()
```

```python
df.describe()
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split

scaler = StandardScaler()
X = scaler.fit_transform(df.drop('Class', axis=1).values)
y = df['Class'].values

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1337)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report


logreg = LogisticRegression(max_iter=1000, multi_class='multinomial')
logreg.fit(X_train, y_train)
y_hat = logreg.predict(X_test)
print(classification_report(y_hat, y_test))
```

```python
from sklearn.model_selection import RandomizedSearchCV

param_grid = {'C': np.random.uniform(0,2, 100)}
clf = RandomizedSearchCV(logreg, param_grid, n_iter=10)
```

```python
search = clf.fit(X_train, y_train)
y_hat = search.predict(X_test)
print(search.best_params_)
print(classification_report(y_hat, y_test))
```
